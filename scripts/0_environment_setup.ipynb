# Installations
Using PIP to install packages.
Python versions were tested successfully from 3.8 to 3.12.

```python
### Python Libraries ###
# Install core libraries
pip install numpy
pip install matplotlib
pip install scipy
pip install scikit-learn
pip install seaborn
pip install umap-learn

# Install file and path handling libraries
pip install natsort

# Install GUI-related libraries
pip install tkinter

# Install ANSYS-related libraries
pip install pyaedt

# Install any additional libraries
pip install pysensors
pip install opencv-python
```
